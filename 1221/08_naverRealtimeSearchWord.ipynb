```python
request = requests.get(targetSite, headers = header)
word = request.text
print(type(word))
print(word)
```

```python
# json       json  import .
import json
```

```python
# json  loads()   json  ()  .
searchWords = json.loads(word)
print(type(searchWords))
print(searchWords)
```

```python
#    .
ranks = searchWords['data']
print(type(ranks))
print(ranks)
```

```python
#           .
for rank in ranks:
    print('{0:2d} => {1} {2}'.format(rank['rank'], rank['keyword'], \ 
                                      rank['keyword_synonyms'] if len(rank['keyword_synonyms']) != 0 else ''))
```
